```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
```

```python
df = pd.DataFrame(columns=['title','content','type'])
df
```

```python
sections = ['business','sports','political-pulse','entertainment','lifestyle','education']


for sec in sections:
  url = 'https://indianexpress.com/section/'+sec+'/'
  page = requests.get(url)
  soup = bs(page.content,'html.parser')
  titles = soup.find_all('div', class_='img-context')
  for title in titles:
      link = title.find('a')
      if link:
          href = link.get('href')
          title_text = link.get('title')
          news_page = requests.get(href)
          news_soup = bs(news_page.content,'html.parser')
          story_details_div = news_soup.find('div', id='pcl-full-content', class_='story_details')
          all_p_tags = story_details_div.find_all('p')
          content = ' '.join([p.get_text(strip=True) for p in all_p_tags])
          df.loc[len(df)] = [title_text, content, sec]
```
